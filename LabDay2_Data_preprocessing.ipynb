## Missing value

```
data = data.replace('?',np.nan)
print(data.isna().sum(axis=0))
row,col =data.shape
data=data.dropna(thresh=row/2,axis=1)
data=data.dropna(thresh=col/3,axis=0)
data.describe(include=include)
```

```
data.isna().sum(axis=0)
#data.isna().sum(axis=1)
```

```
# 
data.A2=data.A2.astype(float)
data.A14=data.A14.astype(float)
```

```
#  mode  mean 
theDataMean = data.mean()
theDataMode = data.mode()
modeNmean=theDataMode
modeNmean[theDataMean.index.values.tolist()]=theDataMean.values.tolist()
modeNmean
```

```
# fill missing value with mode and mean
data.fillna(modeNmean)
data
```

## Transformation

```
catalog_feature_mask=data.dtypes==object
print('catalog_feature_mask')
print(catalog_feature_mask)
categorical_cols=data.columns[catalog_feature_mask].tolist()
print('categorical_cols')
print(categorical_cols)
numerical_cols=data.columns[[not x for x in catalog_feature_mask]].tolist()
print('numerical_cols')
print(numerical_cols)
```

```
# One hot encoder in pandas
cleaned_data=data.drop(categorical_cols,axis=1)
cleaned_data=cleaned_data.join(pd.get_dummies(data[categorical_cols]))
cleaned_data
```

# Outlier

```
Q1 = cleaned_data[numerical_cols].quantile(0.25)
Q3 = cleaned_data[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```
outlier_index=((cleaned_data[numerical_cols] < (Q1 - 1.5 * IQR)) |(cleaned_data[numerical_cols] > (Q3 + 1.5 * IQR)))
```

```
outlier_index.any(axis=1).value_counts().plot.bar()
```

```
outlier_index.all(axis=1).value_counts().plot.bar()
```

```
cleaned_data=cleaned_data[outlier_index.any(axis=1)]
```

```
cleaned_data
```

```
from sklearn.preprocessing import StandardScaler
# Separating out the features
x = cleaned_data[numerical_cols].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```
np.any(np.isnan(cleaned_data[numerical_cols].values))
```

```
scaled_data=pd.DataFrame(data=x,columns=numerical_cols,index=cleaned_data.index)
scaled_data
```

```
scaled_data
```

## PCA

```
scaled_data=scaled_data[np.logical_not(np.isnan(scaled_data).any(axis=1))]
```

```
plt.plot(scaled_data.A2, scaled_data.A3, 'o')
plt.axis('equal');
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_trans = pca.fit_transform(scaled_data[['A2','A3']].values)
plt.plot(X_trans[:, 0], X_trans[:, 1], 'ob', alpha=0.8)
plt.axis('equal');
```

```

pca = PCA(n_components=2)
X_trans=pca.fit(scaled_data.values)
print(pca.explained_variance_)
print(pca.components_)
```

```
# PCA with 1 component
model = PCA(n_components=1)
X_trans = model.fit_transform(scaled_data.values)          # Project onto principal directions
X_new = model.inverse_transform(X_trans)  # Return to unprojected space
plt.plot(scaled_data.A2, scaled_data.A3, 'o', alpha=0.2)
plt.plot(X_new[:, 0], X_new[:, 1], 'ob', alpha=0.8)
plt.axis('equal');
```
